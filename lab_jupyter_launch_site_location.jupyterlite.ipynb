```python
## Task 1: Mark all launch sites on a map
def create_launch_map(launch_sites_df, map_center):
    """
    Create a Folium map with markers for launch sites.

    Parameters:
    - launch_sites_df (pd.DataFrame): DataFrame containing launch site data.
    - map_center (list): Coordinates for the center of the map.

    Returns:
    - folium.Map: Folium map object with markers.
    """
    launch_map = folium.Map(location=map_center, zoom_start=4)

    # Add markers for each launch site
    for index, site in launch_sites_df.iterrows():
        add_launch_marker(launch_map, site)

    return launch_map

def add_launch_marker(launch_map, site):
    """
    Add a marker for a launch site to the given Folium map.

    Parameters:
    - launch_map (folium.Map): Folium map object.
    - site (pd.Series): Series containing launch site data.
    """
    folium.Marker(
        location=[site['Latitude'], site['Longitude']],
        popup=site['Launch Site'],
        icon=folium.Icon(color='darkblue', icon='rocket')
    ).add_to(launch_map)

# Launch site data
launch_sites_data = {
    'Launch Site': ['Site A', 'Site B', 'Site C'],
    'Latitude': [28.5721, 34.0522, 40.7128],
    'Longitude': [-80.6480, -118.2437, -74.0060]
}

# Create DataFrame
launch_sites_df = pd.DataFrame(launch_sites_data)

# Map center coordinates
map_center = [30.0, -95.0]  # Adjust as needed

# Create and display the map
launch_map = create_launch_map(launch_sites_df, map_center)
launch_map
```

```python
import folium
from folium.plugins import MarkerCluster

def create_marker_color(row):
    """
    Assign marker color based on the 'class' column.

    Parameters:
    - row (pd.Series): DataFrame row.

    Returns:
    - str: Marker color ('green' for success, 'red' for failure).
    """
    return 'green' if row['class'] == 1 else 'red'

def create_launch_marker(row):
    """
    Create a folium Marker object based on DataFrame row.

    Parameters:
    - row (pd.Series): DataFrame row.

    Returns:
    - folium.Marker: Folium Marker object.
    """
    return folium.Marker(
        location=[row['Lat'], row['Long']],
        icon=folium.Icon(color='white', icon_color=row['marker_color']),
        popup=f"Launch Site: {row['Launch Site']}<br>Outcome: {'Successful' if row['class'] == 1 else 'Failed'}"
    )

# Add marker color column
spacex_df['marker_color'] = spacex_df.apply(create_marker_color, axis=1)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Create a folium Map object
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
site_map = folium.Map(location=map_center, zoom_start=5)

# Add marker_cluster to the current site_map
site_map.add_child(marker_cluster)

# Add markers to the marker_cluster
for index, row in spacex_df.iterrows():
    marker = create_launch_marker(row)
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
